```python
 %matplotlib notebook
from __future__ import division
import numpy as np
import adios as ad
import matplotlib.pyplot as plt
```

```python
fmesh = ad.file('xgc.mesh.bp')
rz = fmesh['/coordinates/values'][:]
conn = fmesh['/cell_set[0]/node_connect_list'][:]
R = rz[:,0]
Z = rz[:,1]
```

```python
plt.figure(2)
plt.triplot(R, Z, conn);
```

```python
f = ad.file('xgc.3d.08000.bp')
## '0' plane only
dpot = f['dpot'][:,0]
```

```python
plt.figure(10)
plt.tricontourf(R, Z, conn, dpot,cmap=plt.cm.jet, levels=np.linspace(-75,70,num=250));
plt.colorbar();
```
