```python
f = ad.file('xgc.3d.08000.bp')
## '0' plane only
dpot = f['dpot'][:,0]
fmesh = ad.file('xgc.mesh.bp')
rz = fmesh['/coordinates/values'][:]
conn = fmesh['/cell_set[0]/node_connect_list'][:]
R = rz[:,0]
Z = rz[:,1]
```

```python
plt.figure(10, figsize=(8,6))
plt.rc('xtick', labelsize=18)          # fontsize of the tick labels
plt.rc('ytick', labelsize=18)   

axis_font = {'fontname':'Arial', 'size':'18'}

plt.xlabel('R', **axis_font)
plt.ylabel('Z',**axis_font )
plt.title('dpot', **axis_font)
plt.tricontourf(R, Z, conn, dpot,cmap=plt.cm.jet, levels=np.linspace(-76,71,num=250));
plt.colorbar();
```
